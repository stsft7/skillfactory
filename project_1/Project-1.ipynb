```python
'''
#ваш код здесь
import re
# Присваиваем значение 'False' потенциально отсутствующим данным о командировках
hh_database['Город, переезд, командировки'] = hh_database['Город, переезд, командировки'].fillna('не готов к командировкам')

# Определяем список городов-миллионников
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 
    'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 
    'Пермь', 'Воронеж', 'Волгоград'
]

# Обрабатываем данные, разделяя признаки
hh_database['Город'] = hh_database['Город, переезд, командировки'].apply(lambda x: re.split(' , ', x)[0])
hh_database['Город'] = hh_database['Город'].apply(lambda x: x.strip() if x.strip() in ['Москва', 'Санкт-Петербург'] else ('город-миллионник' if x.strip() in million_cities else 'другие'))

hh_database['Готовность к переезду'] = hh_database['Город, переезд, командировки'].apply(lambda x: 'готов к переезду' in x)
hh_database['Готовность к командировкам'] = hh_database['Город, переезд, командировки'].apply(lambda x: 'готов к командировкам' in x)

# Удаляем исходный столбец
hh_database.drop(columns=['Город, переезд, командировки'], inplace=True)
hh_database.head(2)
'''
# Создадим функцию для содания признака "Город"
def get_city(arg):
    # Создадим списки из городов-милионников и двух главных городов нашей страны
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 
                      'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    main_cities = ['Москва', 'Санкт-Петербург']
    
    city = arg.split(sep=' , ')[0]                          # Вычленяем город из списка подстрок
    if city in main_cities:                                 # Если город есть в списке главных городов:
        return city                                         # возвращается его название
    elif city in million_cities:                            # Если город есть в списке городов-миллионников:
        return 'город-миллионник'                           # возвращается категория 'город-миллионник'
    else:                                                   # В оставшемся случае:
        return 'другие'                                     # возвращается категория 'другие' 
    
# Создадим функцию для содания признака "Готовность к переезду"
def get_mobility(arg):
    for i in arg.split(sep=' , '):                          # Перебираем разделенные подстроки:
        if 'перее' in i:                                    # Если значение "перее" есть в подстроке с индексом i:   
            if 'не ' in i:                                  # если значение "не " есть в этой же подстроке с индексом i:
                return False                                # возвращается False
            else:                                           # В обратном случае:                    
                return True                                 # возвращается True                     

# Создадим функцию для содания признака "Готовность к командировкам"
def get_business_trip(arg):                                 
    for i in arg.split(sep=' , '):                          # Перебираем разделенные подстроки:
        if 'командировк' in i:                              # Если значение "командировк" есть в подстроке с индексом i:
            if 'не ' in i:                                  # если значение "не " есть в этой же подстроке с индексом i:
                return False                                # возвращается False
            else:                                           # В обратном случае:
                return True                                 # возвращается True
    return False                                            # В случае отсутствия значения "командировк" в списке подстрок возвращается False

# Создаем новые признаки, вызывая функции для получения категорий городов, признаков готовности к переезду и командировкам
hh_database['Город'] = hh_database['Город, переезд, командировки'].apply(get_city) 
hh_database['Готовность к переезду'] = hh_database['Город, переезд, командировки'].apply(get_mobility) 
hh_database['Готовность к командировкам'] = hh_database['Город, переезд, командировки'].apply(get_business_trip) 

```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
'''
#ваш код здесь
#  Импортируем таблицу с курсами валют
df_mdf = pd.read_csv('C:\SF\project_1_1\ExchangeRates.csv')
#  Удалим лтшние столбцы из таблицы с курсами валют
df_mdf.drop(columns=['per', 'time', 'vol'], axis=1, inplace=True)
#  В признаке 'date' изменим тип данных на Datatime
df_mdf['date'] = pd.to_datetime(df_mdf['date'], errors='coerce').dt.date
# df_mdf.loc[(df_mdf['currency'] == 'KZT')]

dframe = hh_database.copy()
df_zp = pd.DataFrame()
# Создадим новый признак (date) и из dframe['Обновление резюме'] преобразуем в дату и время
# и аксессором возбмём только дату
df_zp['date'] = pd.to_datetime(dframe['Обновление резюме'], errors='coerce').dt.date

#  Извлекём подстроку сумма желаемой зарплаты
#  Изменим тип данных в признаке (Зарплата) на int64
df_zp['Зарплата']= dframe['ЗП'].apply(lambda x: str(x).split()[0])
df_zp['Зарплата'] = df_zp['Зарплата'].astype('int64')

#  dframe['ЗП'] тип данных строка, это сумма и валюта.
#  Извлекём подстроку вылюту и поместим в признак df_zp['currency'] нового датафрейма
df_zp['currency']= dframe['ЗП'].apply(lambda x: str(x).split()[1])

# В признаке (currency) наименование валюты приведем к одному стандарту согласно ISO
df_zp['currency'].replace({'руб.':'RUB', 'бел.руб.':'BYN', 'грн.':'UAH', 'сум':'UZS'}, inplace=True)

merge_df = pd.merge(df_zp, df_mdf, on=['date', 'currency'], how='left')
merge_df.to_csv('merge_df.csv', index=False)


merge_df['close'].fillna(1, inplace=True)
merge_df['proportion'].fillna(1, inplace=True)


merge_df['ЗП(руб.)'] = round(merge_df['Зарплата'] * merge_df['close'] / merge_df['proportion'])

merge_df['ЗП(руб.)'].median()
merge_df.head(2)
'''
# Считаем новый датафрейм с курсами валют
exch_rates = pd.read_csv('ExchangeRates.csv', sep=',')

# Шаг 1
# # Переведем признак "Обновление резюме" из таблицы с резюме в формат datetime и достанем из него дату. 
hh_database['Обновление резюме']=pd.to_datetime(hh_database['Обновление резюме'], dayfirst=True).dt.date

# В тот же формат приведем признак "date" из таблицы с валютами.
exch_rates['date'] = pd.to_datetime(exch_rates['date'], dayfirst=True).dt.date

# Шаг 2
# При помощи laмbda-функции выделим из столбца "ЗП" сумму желаемой заработной платы и 
# наименование валюты, в которой она исчисляется. 
hh_database['Сумма'] = hh_database['ЗП'].apply(lambda x: int(x.split()[0])) 
hh_database['Валюта'] = hh_database['ЗП'].apply(lambda x: x.split()[1])

# Наименование валюты переведем в стандарт ISO.
def iso_rename(arg):
    if 'бел' in arg:
        return 'BYN'
    elif 'грн' in arg:
        return 'UAH'
    elif 'сум' in arg:  
        return 'UZS'
    else:
        return arg
    
hh_database['Валюта'] = hh_database['Валюта'].apply(iso_rename)  

# Шаг 3
# Присоединяем к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты. 
hh_database = hh_database.merge(
    exch_rates, 
    how='left', 
    left_on=['Обновление резюме', 'Валюта'], 
    right_on=['date', 'currency']
)
# Значения close и proportion со значением NaN заполним единицей 1
hh_database['close'] = hh_database['close'].fillna(1)
hh_database['proportion'] = hh_database['proportion'].fillna(1)

# Шаг 4
# Умножим сумму желаемой заработной платы на присоединенный курс валюты (close) и разделим на пропорцию (proportion), 
# результат занесем в новый столбец "ЗП (руб)".
hh_database['ЗП (руб)'] = hh_database['Сумма'] *hh_database['close'] / hh_database['proportion']

# Удалим исходные столбцы : 'ЗП', 'Сумма', 'Валюта' и все столбцы из присоединенной таблицы с валютами
hh_database=hh_database.drop(columns=['ЗП', 'Сумма', 'Валюта', 'currency', 'per', 'date', 'time', 'close', 'vol','proportion'], axis=1)

# Выполним контрольное задание
print("Задание 3.6. Желаемая медианная заработная плата соискателей составляет", round(hh_database['ЗП (руб)'].median()/1000), "тысяч рублей")

# Сравним полученный датафрейм с представленным на платформе для сверки
display(hh_database.info()) # Результат сравнения - положительный
```
